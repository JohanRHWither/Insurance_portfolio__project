```python
ages = []
genders = []
bmis = []
children = []
smokers = []
regions = []
charges = []
#Lists that will contain information on each column
```

```python
with open('insurance.csv') as insurances:
    insure = csv.DictReader(insurances)
    for row in insure:
        ages.append(row['age'])
        genders.append(row['sex'])
        bmis.append(row['bmi'])
        children.append(row['children'])
        smokers.append(row['smoker'])
        regions.append(row['region'])
        charges.append(row['charges'])
```

```python
higher_93 = insurance[insurance['charges']>=39726]
higher_93 = higher_93.sort_values(by = 'charges',ascending=False)
higher_93
```
